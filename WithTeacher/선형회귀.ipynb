```python
import numpy as np
score = np.array([100,90,95,70])
happy = np.array([10,8,9,5])
w = 1/20
a = 0.000001
for i in range(100):
  b=w
  for j in range(4):
    w = w-a*(2*w*score[j]*score[j]-2*score[j]*happy[j])
    if b==w:
      break
print(w)
```
